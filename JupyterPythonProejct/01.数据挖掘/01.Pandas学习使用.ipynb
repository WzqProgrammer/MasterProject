```python
def tanh(x):
    return np.sinh(x)/np.cosh(x)
```

```python
def Sigmoid(x):
    return 1. / (1 + np.exp(-x))
```

```python
filename = 'big_data/all_ratting.json'
with open(filename, 'r', encoding='utf-8') as fr:       
    all_ratting = json.load(fp=fr)
```

```python
test_list = random.sample(all_ratting, 400)
```

```python
df_test = pd.DataFrame(test_list, columns=['uid', 'sid', 'subscribedCount', 'playCount']) 
```

```python
df_test
```

```python
df_test['total'] = df_test['subscribedCount'] * 55 + df_test['playCount']
```

```python
df_test
```

```python
total_min = df_test.total.min()
total_max = df_test.total.max()
total_s = total_max - total_min
```

```python
total_s 
```

计算最值归一化

```python
df_test['extream_normal'] = (df_test['total'] - total_min) / (total_s)
```

计算均方差归一化

```python
total_mean = df_test.total.mean()
total_std = df_test.total.std()
```

```python
total_mean
```

```python
total_std
```

```python
df_test['std_normal'] = (df_test['total'] - total_mean) / (total_std)
```

```python
df_test
```

```python
df_test['tanh_extream'] = df_test['extream_normal'].map(Sigmoid)
```

```python
df_test
```

```python
Sigmoid(0.039425)
```

```python
df_test.rename(columns={'extream_normal': 'extreme_normal'}, inplace=True)
```

```python
df_test.rename(columns={'tanh_extream': 'tanh_extreme'}, inplace=True)
```

```python
df_test
```

```python
df_test['tanh_extreme'] = df_test['extreme_normal'].map(tanh)
```
