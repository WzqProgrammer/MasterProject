```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
```

```python
file = 'data.xlsx'
df = pd.read_excel(file)
```

```python
df.head()
```

```python
matplotlib.rcParams['font.sans-serif']=['SimHei']
matplotlib.rcParams['axes.unicode_minus']=False
```
