# 数据合并

```python
# 初始化数据库连接，使用pymysql模块
# MySQL的用户：root, 密码:147369, 端口：3306,数据库：mydb
engine = create_engine('mysql+pymysql://root:123456@159.75.89.42:3306/easy_call_db')
```

```python
# 根据表名从数据库中读取表数据
def read_data_sql(table_name):
    str_sql = "select * from " + table_name + ";"
    df_res = pd.read_sql_query(str_sql, engine)
    return df_res
```

```python
# 当前项目已有的数据，按获得日期命名
table_names = ['2021_1_06', '2021_1_25', '2021_2_26', '2021_5_19']
```

读取已有的数据

```python
df_1_25 = read_data_sql('2021_1_25')
```

```python
df_5_19 = read_data_sql('2021_5_19')
```

```python
df_list = [df_1_06, df_1_25, df_2_26, df_5_19]
```

```python
tags_map = {}
# 统计当前所有表的标签各自的数量
for df_one in df_list:
    for colume_name in df_one.columns:
        if colume_name not in tags_map.keys():
            tags_map[colume_name] = 1
        else:
            tags_map[colume_name] += 1
```

```python
tags_map
```

```python
tags_4_list = []
tags_3_list = []
for key, value in tags_map.items():
    if value == 4:
        tags_4_list.append(key)
    if value == 3:
        tags_3_list.append(key)
```
