```python
# 需要用众数处理的列标签
mode_tags = ['VIP_LEVEL', 'GENDER', 'IF_BLACK', 'IF_GROUP', 'IF_SCHOOL', 'IF_BRD',
             'IF_DBL_TRMNL', 'IF_4G_TRMNL', 'IF_VOLTE', 'CARD_3G_F', 'IF_ITV']
```

```python
# 需要取保整数的列标签
round_tags = ['VIP_LEVEL', 'ONLINE_MONTH', 'GENDER', 'AGE_SEG', 'TRMNL_AGE', 'BRD_ACTIVE_DAYS', 
             'SMS_COUNT', 'CALL_COUNT']
```

```python
# 不用处理的标签
jump_tags = ['PROD_ID', 'CALL_TIME', 'IF_TOUCH']
```

```python
df_map = {"1_06": df_1_06,"1_25": df_1_25, "2_26": df_2_26, "5_19": df_5_19}
```

```python
for colume in df_1_06.columns:
    if colume in jump_tags:
        continue
    if colume in mode_tags:
        # 众数填充空值，默认取第一个众数
        df_1_06.fillna({colume: df_1_06[colume].mode()[0]}, inplace=True)
    else:
        df_1_06.fillna({colume: df_1_06[colume].mean()}, inplace=True)
    if colume in round_tags:
        df_1_06[colume] = df_1_06[colume].astype(int)
```

```python
# 1_06的VIP_LEVEL的99为空值
df_1_06.loc[df_1_06['VIP_LEVEL'] == 99, 'VIP_LEVEL'] = np.nan
```

```python
def fill_na(df_list):
    for colume in df_list.columns:
        if colume in jump_tags:
            continue
        if colume in mode_tags:
            # 众数填充空值，默认取第一个众数
            df_list.fillna({colume: df_list[colume].mode()[0]}, inplace=True)
        else:
            df_list.fillna({colume: df_list[colume].mean()}, inplace=True)
        if colume in round_tags:
            df_list[colume] = df_list[colume].astype(int)
```
